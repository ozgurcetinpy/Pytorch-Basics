# Reshaping, viewing, stacking tensors

```python
x  = torch.arange(1.,10.)
x, x.shape
```

```python
x_reshaped = x.reshape((3,3))
x_reshaped,  x_reshaped.shape
```

```python
z = x.view(1,9)
z,  z.shape
# changing z changes x cuz a view of a tensor shares the same memory
```

```python
z[:,0] = 5
z, x 
# these are equal
```

```python
x_stacked = torch.stack([x,x,x,x],dim = 0)
x_stacked, x_stacked.shape
```

# Squeezing, unsqueezing and permuting

```python
x = torch.zeros(2,1,2,1,2)
print(x.size())
y = torch.squeeze(x)  # it removes all 1 dimensional tensors
print(y.size())
y = torch.squeeze(x,0)
print(y.size())
y = torch.squeeze(x,1)
print(y.size())
```

```python
x = torch.randn(2,3,5)
print(x.size())
torch.permute(x,(2,0,1)).size()
# second dim is first, zeros dim is in middle...
```

```python
x_original = torch.rand(size = (224,224,3))
x_permuted = x_original.permute(2,0,1)
x_permuted.shape
```

# Reproducibility

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
torch.manual_seed(42)
random_tensor_C = torch.rand(3,4)
torch.manual_seed(42)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D) 
```
